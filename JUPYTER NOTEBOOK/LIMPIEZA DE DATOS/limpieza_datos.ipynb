#### Eliminaci√≥n de columnas y reemplazo de valores

Vemos que la columna fecha_corte es repetitiva y no nos aporta nada, por lo que la eliminamos

##### ITEM

```python
#vemos que la columna ITEM no aporta informacion relevante
df = df.drop(['ITEM'], axis=1)
```

##### FECHA_CORTE

```python
#eliminamos las columna fecha_corte no es necesaria
df = df.drop(['FECHA_CORTE'], axis=1)
```

##### PROGRAMA

Como se puede apreciar a continuacion, la columna PROGRAMA cuenta con solo 2 valores que al final son lo mismo, por lo que se decide ser reemplzada por el string
"PNSR"

```python
#vemos los valores de la columna PROGRAMA
df['PROGRAMA'].value_counts()
```

```python
#reemplazamos los valores de la columna PROGRAMA por "PNSR"
df['PROGRAMA'] = df['PROGRAMA'].replace(['PROGRAMA NACIONAL DE SANEAMIENTO RURAL - PNSR'], 'PNSR')
```

##### PAIS

```python
#eliminaci√≥n de la columna PAIS
df = df.drop(['CODIGO_PAIS'], axis=1)
```

##### NOMBRE_PROYECTO, UNIDAD EJECUTORA, DEPARTAMENTO, PROVINCIA, DISTRITO, MODALIDAD_FINANCIAMIENTO,	ETAPA,ESTADO Y ESTADO_PIP_MEF

```python
#CONVERTIR LAS COLUMNAS NOMBRE_PROYECTO, UNIDAD EJECUTORA, DEPARTAMENTO, PROVINCIA, DISTRITO, MODALIDAD_FINANCIAMIENTO,	ETAPA,ESTADO ,ESTADO_PIP_MEF A MAYUSCULAS
df['NOMBRE_PROYECTO'] = df['NOMBRE_PROYECTO'].str.upper()
df['UNIDAD_EJECUTORA'] = df['UNIDAD_EJECUTORA'].str.upper()
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].str.upper()
df['PROVINCIA'] = df['PROVINCIA'].str.upper()
df['DISTRITO'] = df['DISTRITO'].str.upper()
df['MODALIDAD_FINANCIAMIENTO'] = df['MODALIDAD_FINANCIAMIENTO'].str.upper()
df['ETAPA'] = df['ETAPA'].str.upper()
df['ESTADO'] = df['ESTADO'].str.upper()
df['ESTADO_PIP_MEF'] = df['ESTADO_PIP_MEF'].str.upper()
```

##### FECHA_INICIO y FECHA_FIN

```python
#hacemos una funcion que convierta los valores de la columnas FECHA_INICIO y FECHA_FIN a formato fecha dependiendo de su formato
def convertir_fecha(fecha):
    if type(fecha) == str:
        if fecha.find('/') != -1:
            return pd.to_datetime(fecha, format='%Y-%m-%d').date()
        elif fecha.find('-') != -1:
            return pd.to_datetime(fecha, format='%Y-%m-%d').date()
    elif type(fecha) == int:
        return pd.to_datetime(fecha, format='%Y%m%d').date()
    elif type(fecha) == float:
        return pd.to_datetime(fecha, format='%Y%m%d').date()
    elif type(fecha) == pd._libs.tslibs.timestamps.Timestamp:
        return fecha.date()
    else:
        return fecha
```

```python
#aplicamos la funcion a las columnas FECHA_INICIO y FECHA_FIN
df['FECHA_INICIO'] = df['FECHA_INICIO'].apply(convertir_fecha)
df['FECHA_FIN'] = df['FECHA_FIN'].apply(convertir_fecha)
```

##### UBIGEO
El ubigeo es un codigo que identifica a cada departamento, provincia y distrito del Peru, por lo que se debe ser entero

```python
#convertimos la columna UBIGEO a entero
df['UBIGEO'] = df['UBIGEO'].astype(int)
```

##### CODIGO_SNIP, CODIGO_UNIFICADO

Ambas al ser columnas de codigos se deben ser enteros, por lo que se procede a convertirlos

```python
#convertimos las columnas CODIGO_SNIP y CODIGO_UNIFICADO a entero
df['CODIGO_SNIP'] = df['CODIGO_SNIP'].astype(int)
df['CODIGO_UNIFICADO'] = df['CODIGO_UNIFICADO'].astype(int)
```

##### BENEFICIARIOS

La columna beneficiarios nos indica la cantidad de beneficiarios que se veran afectados por el proyecto, por lo que se debe ser entero

```python
#convertimos la columna BENEFICIARIOS a entero
df['BENEFICIARIOS'] = df['BENEFICIARIOS'].astype(int)
```
